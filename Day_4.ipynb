```python
import pandas as pd
import numpy as np
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('/content/u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("/content/Movie_Id_Titles")
movie_titles.head()
```

```python
#merge together of two dataframes
df = pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
#EDA

#Let's create a ratings dataframe with average rating and number of ratings:
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head(10)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating')
moviemat
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
# now you are watching startwars which is a sci-fiction
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
# we have filtering the movies with ratings >100
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head(10)
```
